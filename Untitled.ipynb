```python
import sqlite3

conn = sqlite3.connect("chroma_aug_store/chroma.sqlite3")
cur = conn.cursor()

# list tables
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("Tables:", cur.fetchall())

# list collections
cur.execute("SELECT id, name FROM collections;")
rows = cur.fetchall()

print("Collections:")
for r in rows:
    print(r)
```

```python
from langchain.vectorstores import Chroma
from sentence_transformers import SentenceTransformer

embedding = SentenceTransformer("all-MiniLM-L6-v2")

db = Chroma(
    persist_directory="chroma_aug_store",
    collection_name="DBMS-25",   # or "langchain"
    embedding_function=embedding
)

collection = db._collection
print("Loaded:", collection.name)
```

```python
data = collection.get(include=["embeddings", "metadatas", "documents"])

X = data["embeddings"]
meta = data["metadatas"]
docs = data["documents"]
```

```python
import numpy as np
import umap
import plotly.express as px

X = np.array(X)

X_3d = umap.UMAP(
    n_components=3,
    metric="cosine",
    random_state=42
).fit_transform(X)

fig = px.scatter_3d(
    x=X_3d[:,0],
    y=X_3d[:,1],
    z=X_3d[:,2],
    title="ChromaDB 3D Embedding Space",
    opacity=0.75
)

fig.update_traces(marker=dict(size=5))
fig.update_layout(height=850)
fig.show()
```
