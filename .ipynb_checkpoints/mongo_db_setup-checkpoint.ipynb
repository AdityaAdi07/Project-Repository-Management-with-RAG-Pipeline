```python
from pymongo import MongoClient, ASCENDING
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

MONGO_URI = os.getenv("MONGO_URI")
if not MONGO_URI:
    raise ValueError("âŒ MONGO_URI not found in .env")

client = MongoClient(MONGO_URI)
db = client["rag_project_db"]

print("âœ… Connected to MongoDB Atlas")
```

```python
collections = {
    "faculty": db["Faculty"],
    "students": db["Student"],
    "teams": db["Teams"],
    "admins": db["Admin"],
    "search_history": db["Search_history"],
    "similarity_logs": db["Similarity_logs"]
}
```

```python
collections["faculty"].create_index([("fac_id", ASCENDING)], unique=True)
collections["students"].create_index([("usn", ASCENDING)], unique=True)
collections["teams"].create_index([("team_id", ASCENDING)], unique=True)
collections["admins"].create_index([("admin_id", ASCENDING)], unique=True)
collections["search_history"].create_index([("srch_id", ASCENDING)], unique=True)
collections["similarity_logs"].create_index([("log_id", ASCENDING)], unique=True)

print("âœ… Primary Keys (unique indexes) created successfully")
```

```python
def validate_fk(collection, key, value):
    """Simulate FK constraint by checking if the reference exists."""
    if value is None:
        return True
    return collection.find_one({key: value}) is not None

```

```python

def insert_faculty(faculty):
    collections["faculty"].insert_one(faculty)
    print("ğŸ“Œ Faculty inserted.")

def insert_student(student):
    # FK: student.team_id â†’ Teams.team_id
    if not validate_fk(collections["teams"], "team_id", student.get("team_id")):
        raise ValueError("âŒ Foreign key error: Invalid team_id in Student")
    collections["students"].insert_one(student)
    print("ğŸ“Œ Student inserted.")


def insert_team(team):
    # FK: team.fac_id â†’ Faculty.fac_id
    if not validate_fk(collections["faculty"], "fac_id", team.get("fac_id")):
        raise ValueError("âŒ Foreign key error: Invalid fac_id in Team")
    collections["teams"].insert_one(team)
    print("ğŸ“Œ Team inserted.")


def insert_project(project):
    # FK: project.team_id â†’ Teams.team_id
    if not validate_fk(collections["teams"], "team_id", project.get("team_id")):
        raise ValueError("âŒ Invalid team_id (FK constraint failed)")

    # FK: project.fac_id â†’ Faculty.fac_id
    if not validate_fk(collections["faculty"], "fac_id", project.get("fac_id")):
        raise ValueError("âŒ Invalid fac_id (FK constraint failed)")

    collections["projects"].insert_one(project)
    print("ğŸ“Œ Project inserted.")


def insert_embedding(embedding):
    # FK: embedding.p_id â†’ Projects.p_id
    if not validate_fk(collections["projects"], "p_id", embedding.get("p_id")):
        raise ValueError("âŒ Invalid p_id in Embedding")

    collections["embeddings"].insert_one(embedding)
    print("ğŸ“Œ Embedding inserted.")


def insert_search_history(history):
    # FK: matched project ID exists
    if not validate_fk(collections["projects"], "p_id", history.get("matched_pid")):
        raise ValueError("âŒ Invalid matched_pid in Search History")

    collections["search_history"].insert_one(history)
    print("ğŸ“Œ Search history inserted.")


def insert_similarity_log(log):
    # FK: log.p_id must exist
    if not validate_fk(collections["projects"], "p_id", log.get("p_id")):
        raise ValueError("âŒ Invalid p_id in Similarity Log")

    # FK: matched_pid must exist
    if not validate_fk(collections["projects"], "p_id", log.get("matched_pid")):
        raise ValueError("âŒ Invalid matched_pid in Similarity Log")

    collections["similarity_logs"].insert_one(log)
    print("ğŸ“Œ Similarity log inserted.")
```

```python
print("ğŸ‰ MongoDB schema setup complete with PK + FK validation!")
```

```python
import json
import subprocess
import re
from pymongo import MongoClient
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
MONGO_URI = os.getenv("MONGO_URI")
client = MongoClient(MONGO_URI)
db = client["rag_project_db"]

teams_col = db["teams"]
students_col = db["students"]

print("âœ… Connected to MongoDB")
```

```python
def author_to_usn(auth_id: str):
    main_num = "".join(re.findall(r"\d+", auth_id.split("_")[0]))  # auth001 â†’ 001

    syn_num = ""
    if "_syn" in auth_id:
        syn_num = "".join(re.findall(r"\d+", auth_id.split("_syn")[-1]))  # syn2 â†’ 2

    return f"1RV23IS{main_num}{syn_num}"

```

```python
def generate_team_name():
    prompt = (
        "Generate a unique, creative, never-before-used CSE-themed team name. "
        "Rules:\n"
        "- Strictly 1 or 2 words only\n"
        "- Avoid generic names like 'Cyber Team', 'Data Group', 'Code Squad'\n"
        "- Must be strongly related to Computer Science (AI, ML, Systems, Security, DB, CV, Theory)\n"
        "- Avoid words used in common team names across colleges\n"
        "- No quotes, no punctuation, no numbering\n"
        "Examples of good style: 'NeuroFlux', 'QuantumOps', 'PixelMatrix', 'CipherWorks', 'LogicForge'\n"
        "Now generate ONE such unique name:"
    )

    response = client.chat.completions.create(
        model="llama3.2",
        messages=[
            {"role": "system", "content": "You generate highly unique, creative team names."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=10,
        temperature=1.0  # higher temp = more uniqueness
    )

    name = response.choices[0].message.content.strip()
    return name.replace('"', "").replace("'", "")
```

```python
JSON_PATH = "projects-embedding-augmented.json"
with open(JSON_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)


print(f"ğŸ“¦ Loaded {len(data)} raw entries")
```

```python
groups = {}  # "auto001": ["auth001", "auth001_syn2", ...]

for entry in data:
    pid = entry.get("project_id")
    if not pid:
        continue

    if "_syn" in pid:
        main_pid = pid.split("_syn")[0]
    else:
        main_pid = pid

    if main_pid not in groups:
        groups[main_pid] = set()

    groups[main_pid].add(entry["source"])  # each entry has "source": authorID


print(f"ğŸ§© Found {len(groups)} project groups")

```

```python
for main_pid, authors in groups.items():

    team_id = f"team_{main_pid}"

    # ----------------------------------------
    # TEAM â€” Insert if not exists
    # ----------------------------------------
    if not teams_col.find_one({"team_id": team_id}):
        t_name = generate_team_name()

        teams_col.insert_one({
            "team_id": team_id,
            "t_name": t_name,
            "fac_id": "F000"
        })

        print(f"ğŸ† Team Created: {team_id} â†’ {t_name}")
    else:
        t_name = teams_col.find_one({"team_id": team_id})["t_name"]
        print(f"âœ” Team Exists: {team_id} â†’ {t_name}")

    # ----------------------------------------
    # STUDENTS â€” Insert all authors (lead + syn)
    # ----------------------------------------
    for auth in authors:
        usn = author_to_usn(auth)

        if not students_col.find_one({"usn": usn}):
            students_col.insert_one({
                "usn": usn,
                "s_name": auth,
                "s_mail_id": f"{auth}@college.edu",
                "dept": "ISE",
                "team_id": team_id
            })
            print(f"ğŸ‘¤ Student Added: {auth} â†’ {usn}")

        else:
            print(f"âœ” Student Exists: {auth} â†’ {usn}")


print("\nğŸ‰ Completed: Teams + Students inserted successfully.")
```
